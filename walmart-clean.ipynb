```python
import pandas as pd
feature=pd.read_csv("E:\Artefact\\artefact\Walmart\\features.csv")
feature.head()
```

```python
feature['Date'] = pd.to_datetime(feature['Date'])
feature['year'] =feature['Date'].dt.year.fillna(0).astype("int")   #转化提取年 
feature['month'] = feature['Date'].dt.month.fillna(0).astype("int")  #转化提取月
#feature['year-month'] = feature['year'].map(str) + '-' + feature['month'].map(str) #提取年月
feature
```

```python
stores=pd.read_csv("E:\Artefact\\artefact\Walmart\\stores.csv")
stores.head()
```

```python
merge=pd.merge(feature, stores, on='Store')
merge
```

```python
train=pd.read_csv("E:\Artefact\\artefact\Walmart\\train.csv")
train['Date'] = pd.to_datetime(train['Date'])
train.head()
```

```python
all=pd.merge(merge, train, on=['Store','Date','IsHoliday'],how='outer')
all
```

```python
all.to_csv(path_or_buf="E:\\all.csv",index=False)
```

```python
all=pd.read_csv("E:\Artefact\\artefact\Walmart\\all.csv")
all['Date'] = pd.to_datetime(all['Date'])
all.head()
open_day='2010-02-05'
close_day='2012-11-01'
con1=all['Date']>=open_day
con2=all['Date']<close_day
clean=all[con1&con2]
clean
```

```python
clean.to_csv(path_or_buf="E:\Artefact\\artefact\Walmart\\clean.csv",index=False)
```

```python
import os
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
path = "E:\Artefact\\artefact\Walmart\\clean.csv"   #导入需要的数据，必须csv
df = pd.read_csv(path)  #命名
sns.set_style('whitegrid',{'font.sans-serif':['simhei','Arial']})   #解决Seaborn中文显示问题
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(),cmap="YlGnBu_r",annot = True,cbar = True)     #命名.corr()作相关性分析
plt.rcParams['axes.unicode_minus']=False #显示负号
plt.xticks(rotation=90)
plt.yticks(rotation = 0)
plt.title('相关系数矩阵')
plt.savefig('相关系数.jpg', dpi=300,bbox_inches='tight')
plt.show()
```
